#This notebook is about HMM in datapop.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lhcb_data = pd.read_csv('Data/popularity-728days.csv')
lhcb_data.irow(range(0,3))
```

```python
from datapop.LHCb import DataPreprocessor
data = DataPreprocessor(data=lhcb_data).get_preprocessed_data()
data.irow(range(0,3))
```

```python
from datapop.core import Simulator
sim = Simulator()
test_data = sim.get_period(data=data, period=78, forecast_horizont=26, class_abs_thresholds=[1])
test_data.irow(range(0,3))
```

##HMM for time series prediction.

```python
%%time
from datapop.core import HMMPredict
hmmp = HMMPredict(data=test_data, forecast_horizont=26, last_periods=None, class_abs_thresholds=[0.01])
report = hmmp.get_forecast_report()
```

```python
report.irow(range(0,10))
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(test_data['True_Type'].values, report['Recommended_Type'].values)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr)
plt.show()
```

```python
plt.hist(report['LowPercentile95'][test_data['True_Type'].values==0][report['LowPercentile95']<=100].values, bins=10, color='r', alpha=0.5)
plt.hist(report['LowPercentile95'][test_data['True_Type'].values==1][report['LowPercentile95']<=100].values, bins=10, color='b', alpha=0.5)
plt.xlim(0,100)
plt.show()
```

#HMM for detection of disorder of time series.

```python
from datapop.LHCb import DisorderTimeDetection
dtd = DisorderTimeDetection(data=test_data)
report = dtd.get_report()
```

```python
report.irow(range(0,3))
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(test_data['True_Type'].values, report['Recommended_Type'].values)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr)
plt.show()
print 'ROC AUC = ', roc_auc
```

```python
selection = (report['Ratio']>0)*(report['Ratio']<10)
r = report['Ratio'][selection].values
l = (test_data[selection.values])['True_Type'].values
plt.hist(r[l==0], label='0', bins=10, alpha=0.5)
plt.hist(r[l==1], label='1', bins=10, alpha=0.5)
plt.legend(loc='best')
plt.show()
```
